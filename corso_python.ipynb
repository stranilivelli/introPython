## Espressioni

Un'espressione √® una combinazione di valori, operatori e funzioni che viene valutata per produrre un risultato. Le espressioni possono essere semplici o complesse, e possono coinvolgere numeri, stringhe, booleani, variabili e molto altro.

Gli operatori vengono valutati in un ordine specifico, dettato dalle regole di precedenza degli operatori. Ad esempio, gli operatori di moltiplicazione e divisione vengono valutati prima degli operatori di somma e sottrazione (esattamente come ci hanno spiegato a scuola per la matematica)


*   Precedenze:
- Parentesi
- Operatori unari (+, -, not)
- Moltiplicazione, divisione, modulo (%), divisione intera (//)
- Addizione e sottrazione
- Operatori di confronto (==, !=, <, >, <=, >=)
- Operatori booleani (and, or)
- Operatori di assegnazione (=, +=, -=, *=, /=, %=, //=)



√à possibile cambiare l'ordine di valutazione degli operatori utilizzando le parentesi. In questo modo, gli operatori all'interno delle parentesi vengono valutati prima degli operatori all'esterno delle parentesi


### Operatori di confronto

Elenco degli operatori di confronto in Python:

```
== (uguale a)
!= (diverso da)
< (minore di)
> (maggiore di)
<= (minore o uguale a)
>= (maggiore o uguale a)
```



````python
# Esempio 1: confronto tra numeri
x = 5
y = 3
print(x == y)  # output: False
print(x != y)  # output: True
print(x < y)  # output: False
print(x > y)  # output: True
print(x <= y)  # output: False
print(x >= y)  # output: True

# Esempio 2: confronto tra stringhe
stringa1 = "learnn"
stringa2 = "Learnn"
print(stringa1 == stringa2)  # output: False
print(stringa1 != stringa2)  # output: True
print(stringa1 < stringa2)  # output: True
print(stringa1 > stringa2)  # output: False
````

### Operatori Booleani

| Operatore |	Descrizione| priorit√† |
|---------- |---------- | ---------- |
| not |	Negazione logica | 1 |
| and | Congiunzione logica | 2 |
| or  | Disgiunzione logica | 3 |

### Esempio pratico
Nella valutazione dell'espressione `not A and B or C`

l'ordine di valutazione √® il seguente:

`not A` viene valutato prima di tutto perch√© ha la precedenza pi√π alta.
Se A √® True, l'espressione diventa False, altrimenti diventa True.

La congiunzione logica` and` viene valutata successivamente.

 Se l'operando sinistro (il risultato di not A) √® False, l'intera espressione diventa False e non viene valutato l'operando destro (B). Altrimenti, viene valutato B.


Infine, la disgiunzione logica `or` viene valutata. Se uno dei due operandi √® True, l'intera espressione √® True. Altrimenti, √® False.

*üí° √à sempre possibile cambiare l'ordine di valutazione degli operatori utilizzando le parentesi.*


````python
# Esempio 1: operatore and
a = True
b = False
print(a and b)  # output: False

# Esempio 2: operatore or
c = True
d = False
print(c or d)  # output: True

# Esempio 3: operatore not
e = True
print(not e)  # output: False
````

````python
# Ordine di valutazione
a = True
b = False
c = True
d = False
print((a or b) and (c or not d))  # output: True
````

````python
import requests
import json
import time
# id prodotto di woocommerce
id_prodotto = 21
# Imposta l'incremento di prezzo per ogni offerta e il numero totale di cicli
incremento_di_prezzo = 0.5
num_di_incrementi = 10

secondi_delay = 5

for i in range(0, num_di_incrementi):

    # URL per leggere i dati dei prodotti dal negozio WooCommerce
    url = "https://damianodotto.altervista.org/wp-json/wc/v3/products"

    # Parametri dell'API
    params = {
        "consumer_key": "ck_6992115379057be1e1def9c0956d255fbbeb5653",
        "consumer_secret": "cs_3df831a676ab547ec8bc17c459cf7b9a76a96969"
    }

    # Leggi i dati dei prodotti dal negozio
    response = requests.get(url, params=params)
    products = json.loads(response.text)

    # Aspetta per il tempo di offerta
    time.sleep(secondi_delay)

    # Utilizza una list comprehension per estrarre il prezzo di ogni prodotto
    prezzo = float([product['price'] for product in products if product['id'] == id_prodotto][0])

    # Stampa il prezzo corrente dell'offerta
    print(f"""Offerta corrente: ‚Ç¨{prezzo:.2f}""")

    # Incrementa il prezzo dell'offerta
    prezzo += incremento_di_prezzo

    # Stampa il nuovo prezzo dell'offerta
    print(f"""Nuova offerta: ‚Ç¨{prezzo:.2f}""")

    # Costruisci i dati da inviare all'API per aggiornare il prodotto
    data = {
        "regular_price": str(prezzo)
    }

    # Invia la richiesta PUT all'API per aggiornare il prodotto
    response = requests.put(f"""https://damianodotto.altervista.org/wp-json/wc/v3/products/{id_prodotto}""",
                            params=params,
                            data=json.dumps(data),
                            headers={'Content-Type': 'application/json'})

    # Verifica se la richiesta √® stata eseguita con successo
    if response.status_code == 200:
        print(f"Prezzo del prodotto aggiornato con successo a {prezzo}!")
    else:
        print("Errore durante l'aggiornamento del prezzo del prodotto.")
        print(f"Status code: {response.status_code}, Response: {response.text}")
````
